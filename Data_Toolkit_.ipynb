**Q3.Explain the following terms:**

 **1. The difference in np.array, np.asarray and np.asanyarray.**

 **2. The difference between Deep copy and shallow copy.**

 **Sol.** **np.array():**
 * It creates an array of a given numpy object.
 * It creates a new copy if array is given in it.   


```python
#Create an array
import numpy as np
import pandas as pd

#it will convert the given list into  the array
a=np.array([1,2,3],dtype=float)
a1=np.array(a)
print(a)
```

```python
#saves a copy of an array to new location

import numpy as np
import pandas as pd

a1=np.array([1,2,3],dtype=float)
a2=np.array(a1)

a1 is a2
```

**np.asarray():**
* It also converts the given numpy object into an array.
* It doesnt create a new copy if the given object is already an array.

```python
#Create an array using asarray()
import numpy as np
import pandas as pd

#it will convert the given list into  the array
a=np.asarray([1,2,3],dtype=float)
print(a)
```

```python
#Doest saves a copy of an array to new location if array is given

import numpy as np
import pandas as pd

a1=np.array([1,2,3],dtype=float)
a2=np.asarray(a1)

a1 is a2
```

**np.asanyarray():**
* Similar to np.asarray(), but allows subclasses of ndarray (like masked arrays) to be passed through without conversion.
* It doesnt create a new copy if the given object is already an array.

```python
#Create an array using asarray()
import numpy as np
import pandas as pd

#it will convert the given list into  the array
a=np.asanyarray([1,2,3],dtype=float)
print(a)
```

```python
#Doest saves a copy of an array to new location if array is given

import numpy as np
import pandas as pd

a1=np.array([1,2,3],dtype=float)
a2=np.asanyarray(a1)

a1 is a2
```

```python
#Shalow copy:

import numpy as np
import pandas as pd

a=[[1,2,6,7],[8,9,11,13]]
b=a.copy()

lst=[25,62,35,77]
c=lst.copy()

#NESTED LIST CASE :making change in b
b[0][0]=1000
print("Both a and b will be same after change in b :")
print("A:",a)
print("B:",b)


#NON NESTED LIST CASE: making change in c
c[0]=50000

print("\nBoth a and c will be different after change in c :")
print("A:",a)
print("C:",c)

print("A and C are different because after changing the value in c,it is allocated to a different memory location")

```

```python
#Deep Copy

import numpy as np
import pandas as pd
import copy as cp

a=[[1,2,6,7],[8,9,11,13]]
b=cp.deepcopy(a)

lst=[25,62,35,77]
c=cp.deepcopy(lst)

#NON NESTED LIST CASE:
print("In case of non nested lists shalow copy and nested copy is same")
c[0]=50000

print("\nBoth a and c will be different after change in c :")
print("A:",a)
print("C:",c)

#NESTED LIST CASE :making change in b
b[0][0]=1000
print("\nBoth a and b will be different after change in b :")
print("A:",a)
print("B:",b)

print("\nA and B are different,This is because in case of deep copy when cp.deepcopy() was executed 'b' got a different memory location")
print("Thus changes made in 'b' are no longer gonna affect 'a'")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

x = np.random.randint(-10,10,100)
y = np.random.randint(-10,10,100)
z=[]

def quad(a,b):
  """Checks for the quadrent of the coordinates
         a,b  = int,int
      returns = origin,I,II,III,IV quad
  """
  for i in range(0,100):
    if a[i]>0 and b[i]>0:
      z.append("I")
    elif a[i]<0 and b[i]>0:
      z.append("II")
    elif a[i]<0 and b[i]<0:
      z.append("III")
    elif a[i]>0 and b[i]<0:
      z.append("IV")
    else:
      z.append("Origin")

quad(x,y)

df = pd.DataFrame({'X': x, 'Y': y, 'Quadrant': z})

plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(title='Scatter Plot')

sns.scatterplot(data=df, x='X', y='Y', hue='Quadrant', style='Quadrant')

plt.grid()
plt.show()
```
